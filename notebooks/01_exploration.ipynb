```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/telco_churn_clean.csv")

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df = df.dropna(subset=['TotalCharges'])
# Group and calculate mean tenure
tenure_mean = df.groupby("Churn")["tenure"].mean().reset_index()

# plot
sns.countplot(x='Churn', data=df)
plt.savefig("../reports/figures/churn_distribution.png", dpi=300)
plt.show()

# plot
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.savefig("../reports/figures/monthly_charges_vs_churn.png", dpi=300)
plt.show()

# Plot
plt.figure(figsize=(6, 4))
sns.barplot(x="Churn", y="tenure", data=tenure_mean)
plt.title("Average Tenure by Churn Group")
plt.xlabel("Churn")
plt.ylabel("Average Tenure (months)")
plt.tight_layout()

# Save for report
plt.savefig("../reports/figures/tenure_vs_churn_bar.png", dpi=300)
plt.show()
```
